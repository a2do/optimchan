Đánh giá hiệu suất của một thuật toán tối ưu

- Đánh giá độ lỗi tuyệt đối qua mỗi lần lặp: sai khác tuyệt đối giữa đánh giá hàm mục tiêu bằng cách sử dụng tham số được cho bởi thuật toán tối ưu, gọi là $J(\hat{\theta_t})$ và giá trị tối ưu thật sự, đặt là $J(\theta_{opt})$, tức là khoảng cách tuyệt đối đến điểm tối ưu mà chỉ xem xét trên trục $z$:

$$
    |J(\hat{\theta_z}) - J(\theta_z)|
$$

- Khoảng cách Euclidean đến điểm tối ưu toàn cục của hàm mục tiêu, sử dụng $l2$-norm:

$$
    l_2 = || \{\theta_{opt},J(\theta_{opt})\}, \{\hat{\theta_t},J(\hat{\theta_t})\} ||_2
$$

- Đánh giá về không gian: 
  - Đánh giá theo thời gian với tham số $X$
  - Đánh giá theo thời gian với tham số $Y$
  - Đường đi ngắn nhất đến điểm tối ưu được tính toán bởi thuật toán

- Đánh giá về sự linh động
  - Đánh giá độ lớn của vận tốc trên trục $x$, $|\vec{v_x}|$ theo thời gian
  - Đánh giá độ lớn của vận tốc trên trục $y$, $|\vec{v_y}|$ theo thời gian
  - Đánh giá độ lớn của vận tốc $|\vec{v}|$ theo thời gian
